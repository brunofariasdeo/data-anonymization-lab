## 4. Técnicas de anonimização de dados

### 4.1. Generalização de dados

#### A generalização reduz a granularidade dos dados para evitar a identificação. Isso permite que os dados permaneçam úteis para análise, reduzindo o risco de reidentificação.

#### Por exemplo, ao lidar com dados pessoais como data de nascimento, em vez de mostrar a data exata de nascimento, os dados são generalizados para o mês e ano, ou apenas o ano, para evitar a identificação, mantendo a faixa etária relevante para análise. 

#### Exemplo

```python
df['Age'] = pd.to_numeric(df['Age'])                       
df['Age'] = pd.cut(df['Age'], bins=[0, 30, 40, 50, 100], labels=['18-30', '31-40', '41-50', '51+'])

display(df)
```

### 4.2. Pertubação de dados

#### A perturbação de dados é uma técnica valiosa quando precisamos proteger a privacidade em análises que não requerem dados individuais precisos, mas sim a distribuição geral dos dados. Este método envolve a modificação controlada dos dados originais, utilizando técnicas como aleatorização, escalonamento ou troca de valores. 

#### Exemplo

```python
df['Billing Amount'] = pd.to_numeric(df['Billing Amount'])
df['Billing Amount'] = df['Billing Amount'] + random.uniform(0, 100000)

display(df)
```

### 4.3. Mascaramento de dados

#### O mascaramento de dados é uma técnica amplamente utilizada para proteger informações sensíveis. Esse método envolve alterar ou ofuscar os dados originais, mantendo seu formato e estrutura.

#### Exemplo

```python
df['Name'] = df['Name'].apply(lambda x: 'Pessoa_' + ''.join(np.random.choice(list('ABCDEFGHIJKLMNOPQRSTUVWXYZ'), 5)))

display(df)
```
